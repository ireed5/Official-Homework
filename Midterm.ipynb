```python
#read the file numbers.txt
f = open("numbers.txt", "r")
num = f.read()
num1 = num.replace('\n', ',').split(',')
num2 = [int(i) for i in num1]
num2.sort()
print(num2)
```

```python
#find the frequency of each distinct element using a dictionary
frequency = {}
for number in num2:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1
print(frequency)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#plot the frequency results in a histgram
num2 = [key for key, val in frequency.items() for _ in range(val)]
plt.hist(num2, bins= 11, color = 'hotpink')
plt.title("frequency of elements")
```

```python
#store dictionary results into json file
import json
y = json.dumps(frequency)
print(y)
```

```python
#monthly summary
df['Month_Number'] = pd.DatetimeIndex(df['Order Date']).month
def floatdollar(x):
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)
df['Total Charged'] = df['Total Charged'].apply(floatdollar).astype('float')
df.groupby('Month_Number')['Total Charged'].describe()
```

```python
df2 = df.groupby('Month_Number')['Total Charged'].describe()
```

```python
#Compare Average Spent per Month Using Bar Chart
ax2 = df2.plot.bar(y='mean', color = 'orange')
plt.xlabel('Month')
plt.ylabel('Dollar Value')
plt.title('Average Spent per Month')
```

```python
#Compare Orders per Month Using Pie Chart
mylabels = ["April", "May", "June", "July"]
mycolors = ["lightpink", "plum", "teal", "cornflowerblue"]
plot = df2.plot.pie(y='count', labels = mylabels, colors = mycolors, title = 'Orders per Month')
plot.legend(title='month', bbox_to_anchor=(1.5,1), loc = 'upper left')
```

```python
#Compare Max and Min each Month Using Multiple Line Comparison
ax = plt.gca()
df2.plot(y='max', ax=ax, color = 'lightpink')
df2.plot(y='min', ax=ax, color = 'orchid')
plt.xlabel('Month')
plt.ylabel('Dollar value')
plt.title('Minimum vs Maximum Spent per Month')
```
