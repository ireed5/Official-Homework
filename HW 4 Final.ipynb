```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#multiple lines
q41 = np.array([1, 6])
q42 = np.array([2, 7])
q43 = np.array([3, 8])
plt.subplot(2,3,1)
plt.plot(q41)
plt.plot(q42)
plt.plot(q43)
plt.title("multiple lines")
plt.show()

#bar chart
x5 = np.array(["An", "Am", "Ja", "Ju", "Ar"])
y5 = np.array([88, 66, 90, 55, 77])
plt.subplot(2,3,2)
plt.bar(x5, y5)
plt.title("bar chart")
plt.show()

#pie chart
y5 = np.array([88, 66, 90, 55, 77])
labels = ["Andy", "Amy", "James", "Jules", "Arthur"]
plt.subplot(2,3,3)
plt.pie(y5, labels = labels)
plt.legend(bbox_to_anchor=(1.2, 1.0), loc = 'upper left')
plt.title("pie chart")
plt.show()

#a grid
q41 = np.array([1, 6])
q42 = np.array([2, 7])
q43 = np.array([3, 8])
plt.subplot(2,3,4)
plt.plot(q41)
plt.plot(q42)
plt.plot(q43)
plt.title("grid")
plt.grid()
plt.show()

#scatterplot
x6 = np.array([2, 3, 4, 5, 6])
y6 = np.array([88, 66, 90, 55, 77])
plt.subplot(2,3,5)
plt.scatter(x6, y6)
plt.title("scatterplot")
plt.show()

#histogram
x7 = np.array([2, 3, 3, 5, 6, 6, 6])
plt.subplot(2,3,6)
plt.hist(x7)
plt.title("histogram")
plt.show()
```
